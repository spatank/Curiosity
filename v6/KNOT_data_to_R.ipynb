```python
import scipy
import scipy.io as sio
from scipy.io import savemat
import pandas as pd
import numpy as np
import math
```

```python
orig_KNOT_df = pd.read_csv('/Volumes/My Passport/Curiosity/Data/KNOT_data_raw.csv')
orig_KNOT_df.head()
```

```python
ID_groups = orig_KNOT_df.groupby('ID')
```

```python
mat_contents = sio.loadmat('/Volumes/My Passport/Curiosity/v6/Data/KNOT/KNOT_data_v2.mat')
ID = mat_contents['ID']
DS = mat_contents['DS_5D']
JE = mat_contents['JE_5D']
SC = mat_contents['SC_5D']
ST = mat_contents['ST_5D']
TS = mat_contents['TS_5D']
density = mat_contents['density']
num_nodes = mat_contents['num_nodes']
num_edges = mat_contents['num_edges']
all_C = mat_contents['all_C']
all_DoF = mat_contents['all_DoF']
all_Betti = mat_contents['all_Betti']
```

```python
df = pd.DataFrame(columns = ['ID', 
                             'Node', 
                             'DS',
                             'JE',
                             'SC',
                             'ST',
                             'TS',
                             'density', 
                             'num_nodes', 
                             'num_edges',
                             'C', 
                             'DoF', 
                             'Betti',
                             'Age',
                             'SexOrient',
                             'Race',
                             'GenderFactor',
                             'EducDeg',
                             'Income'])
for i, subj_ID in enumerate(np.squeeze(ID).tolist()):
    print('Processing subject %d.' % subj_ID)
    subj_C = all_C[i, :]
    subj_DoF = all_DoF[i, :]
    subj_Betti = all_Betti[i, :]
    for j in range(subj_C.shape[0]):
        if math.isnan(subj_C[j]) and math.isnan(subj_DoF[j]) and math.isnan(subj_Betti[j]):
            continue
        df = df.append({'ID': int(subj_ID), 
                        'Node': int(j + 1), 
                        'DS': DS[0][i], 
                        'JE': JE[0][i], 
                        'SC': SC[0][i], 
                        'ST': ST[0][i], 
                        'TS': TS[0][i], 
                        'density': density[0][i], 
                        'num_nodes': num_nodes[0][i], 
                        'num_edges': num_edges[0][i], 
                        'C': subj_C[j], 
                        'DoF': subj_DoF[j], 
                        'Betti': subj_Betti[j], 
                        'Age': ID_groups.get_group(subj_ID).reset_index()['AgeYears'][0], 
                        'SexOrient': ID_groups.get_group(subj_ID).reset_index()['SexOrient'][0], 
                        'Race': ID_groups.get_group(subj_ID).reset_index()['Race'][0], 
                        'GenderFactor': ID_groups.get_group(subj_ID).reset_index()['GenderFactor'][0], 
                        'EducDeg': ID_groups.get_group(subj_ID).reset_index()['EducDeg'][0], 
                        'Income': ID_groups.get_group(subj_ID).reset_index()['Income'][0]},
                       ignore_index = True)
```
